# Data Cleaning

```python
# Counting the total number of duplicates in the Univeristy Ranking Dataframe

uni_df.duplicated().sum()
```

```python
# Dropping all the duplicates rows in the University Ranking Dataframe

uni_df= uni_df.drop_duplicates()
```

```python
#Checking from missing values in the Dataframe "True" means there are missing values and "False" means there are no missing values.

pd.isnull(uni_df)
```

```python
# Counting  how many columns have the missing value null in the Dataframe

pd.isnull(uni_df).sum()
```

```python
#Going through each columns in the University ranking Dataframe
for col in uni_df.columns:

#Checking if the column datatype is object
  if uni_df[col].dtype != "object":
# Filling missing value of the non-object columns with the mean
    uni_df[col] = uni_df[col].fillna(uni_df[col].median())


```

```python
#Dropping rows where the Nae of university columns has missing value
uni_df.dropna(subset =['Name of University'], inplace = True)
```

```python
# Filling mising values in the location column with Unknown
uni_df['Location'].fillna("unknown", inplace = True)
```

```python
# Adding a new column 'University Rank' based on the index of 'Name of university' +1

# Using lambda functions to calculate University rank for each university
uni_df['University Rank'] = uni_df['Name of University'].apply(lambda x : uni_df['Name of University'].tolist().index(x)+1)

# Displaying the first few updated rows.
uni_df.head(1817)
```

```python
#. Seperating the Female and Male ratio column into two seperate individua column by using spilit ":"
uni_df [['Female Ratio','Male Ratio']] = uni_df['Female:Male Ratio'].str.split(':', expand =True)

# Displaying the few rows to see the spilit in the dataframe
uni_df.head()
```

```python
# Convert the Female ratio column into the float data type
uni_df['Female Ratio'] = uni_df['Female Ratio'].astype (float)

#Convert the Male ratio column into the float data type
uni_df['Male Ratio'] = uni_df['Male Ratio'].astype(float)

# Displaying some rows to see the changes in the female and male ratio
uni_df.head()
```

```python
# Extract numeric values with decimal points from the overall score column and convert then to float and
def changeScore(x):
  try:
    return float(x)
  except ValueError:
    return None

#Applying the changeScore function to OverAll Score .
uni_df['OverAll Score'] = uni_df['OverAll Score'].apply (changeScore)
uni_df['OverAll Score'].head(20)


```

```python
# Convert International Student columns to numeric format
uni_df['International Student'] = pd.to_numeric(uni_df['International Student'], errors ="coerce")

# To see the international dtype is changed into numeic format or not
uni_df.info()
```

```python
#Going through each columns in the world university data frame
for col in uni_df.columns:

# Check is the dtype is object if it is replacing missing value by calaculating mode
  if uni_df[col].dtype == "object":
    uni_df[col] = uni_df[col].replace(np.nan,uni_df[col].mode()[0])

# If it is numeric then it is replacing by  calculating  mean
  else:
    uni_df[col] = uni_df[col].replace(np.nan,uni_df[col].median())

#Display the few rows to see missing vlaue is filled or not
uni_df.head()
```

```python
# Conerting the 'No of Student' columns to integer data type
uni_df['No of student'] = uni_df['No of student'].astype(int)

# Displaying the information after converting into integer data type
uni_df.info()

# Displaying few rows to see the integer data type in dataframe
uni_df.head()
```

```python
# Going through each columns from the data frame
for col in uni_df.columns:

# If the column datatype is float  it will round up to 2 .
  if uni_df[col].dtype == "float64":
    uni_df[col]= uni_df[col]. round(2)

# Displaying some rows to see the changes
uni_df.head()

```

```python
# Creating a new columns  to keep the value Multiplying  international student and no of student
uni_df['International Students'] = uni_df['International Student'] * uni_df['No of student']

# Displaying few rows to see the changes in the dataframe
uni_df.head()
```

```python
# Creating a new variable and storing the columns International Student and Female:Male Ratio
columnsDrop = ['International Student', 'Female:Male Ratio']

# Drooping the variable columnsdrop  from the dataframe
uni_df.drop(columns=columnsDrop, axis =1, inplace = True)

# Displaying few rows to see whether the columns are drop or not
uni_df.head()
```

```python
# Filling missing values in International Student columns with zero
uni_df['International Students'] = uni_df['International Students'] .fillna(0,limit=3)

# Convert International students columns into integer datype
uni_df['International Students'] = uni_df['International Students'] .astype(int)

#Displaying few rows to see the changes
uni_df.head()
```
